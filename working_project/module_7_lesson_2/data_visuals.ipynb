```python
#import  libaries
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('anscombe_quartet_data.csv')
print(df)
```

```python
#the correlation coefficient of anscombe's dataset
df.corr()
```

```python
#create the subplots that show how each x column correlates with each y column
#approach one
fig = plt.figure(figsize=(9, 7))
x = [0,15]
y = [0,10.9]
x1 = df['x1']
y1 = df['y1']
plt.subplot(2,2,1)
plt.title('x1,y1')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color ="red")
plt.plot(x,y, color = 'purple', linestyle = 'solid')

x = [0,14]
y = [0,9.32]
x2 = df['x2']
y2 = df['y2']
plt.subplot(2,2,2)
plt.title('x2,y2')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color ="blue")
plt.plot(x,y, color = 'purple', linestyle = 'solid')

x = [0,14]
y = [0,12.7]
x3 = df['x3']
y3 = df['y3']
plt.subplot(2,2,3)
plt.title('x3,y3')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color ="yellow")
plt.plot(x,y, color = 'purple', linestyle = 'solid')

x = [0,19]
y = [0,12.7]
x4 = df['x4']
y4 = df['y4']
plt.subplot(2,2,4)
plt.title('x4,y4')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color ="brown")
plt.plot(x,y, color = 'purple', linestyle = 'solid')

plt.subplots_adjust(wspace=0.5, hspace=0.5, bottom=0.1, top=0.9)
plt.show()
```
