```python
#import libraries
import pandas as pd
```

```python
#creating a dataset called orders
orders = {
    "order_id": ['O1','O2','O3','O4','O5','O6','O7','O8','O9','O10','O11','O12','O13','O14','O15','O16','O17','O18','O19','O20'],
    "customer_id": ['C1','C2','C3','C4','C5','C5','C3','C4','C9','C10','C6','C7','C13','C14','C1','C8','C9','C11','C10','C2'],
    "product_id": ['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15','P16','P17','P18','P19','P20'],
    "quantity": [4,7,9,3,6,1,2,5,8,10,14,18,13,17,16,11,12,15,19,20]
         }

df_order = pd.DataFrame(orders)
(df_order)
```

```python
#creating a dataset called products 
products = {
    "product_id": ['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15'],
    "product_description": ['butter','bread','brownies','bama','oats','milk','tissue','plate','sardine','cornflakes','coco pops','chocolate','biscuits','noodles','spoons'],
    "product_price": [400,2200,3400,500,5000,5400,2500,3600,3200,5700,4500,450,6700,8900,9000]
           }

df_products = pd.DataFrame(products)
print(df_products)
```

```python
#create a dataset called customers
customers = {
    "customer_id": ['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10'],
    "customer_name": ['Gideon','Adams','Esther','Mary','Angela','Sarah','Williams','John','Joy','Chioma'],
    "customer_address": ['Ikotun','Ajah','Surulere','Masha','shitta','Gbagada','Ikeja','Yaba','Sabo','Ikoyi'],
    "customer_phone_number": ['08087654567','08098761234','08022345567','09077754432','09011223344','09045324768','07022335768','07077968765','07033427685','07123454345']
            }

df_customers = pd.DataFrame(customers)
(df_customers)
```

```python
#inner merge operation between the orders table and products table
df_ordersproducts = df_order.merge(df_products, how='inner', indicator = True)
print(df_ordersproducts.head())
```

```python
#another inner merge operation with the customers table using the resulting table
df_all = df_ordersproducts.merge(df_customers, how='inner')
(df_all.head())
```

```python
#inner merge operation between customers and orders tables. on customer_id
df_customersorders = df_customers.merge(df_order, how='inner',left_on ='customer_id', right_on = 'customer_id', suffixes =('_customers','_orders'))
df_customersorders.head()
```

```python
#outer merge between the customers and products and products tables. using left_index, right_index
df_customersproducts = pd.merge(df_customers, df_products, left_index=True, right_index=True, how ='outer', indicator=True)
(df_customersproducts.head())
```

```python
#left merge between orders and customers tables
df_orderscustomers = pd.merge(df_order,df_customers, how='left', on='customer_id')
(df_orderscustomers)
```

```python
# right merge between product and customer tables
df_productscustomers = pd.merge(df_products, df_customers, left_index = True, right_index = True, how = 'right')
df_productscustomers.head()
```

```python
#using the pandas concatenation method to join the three datasets together
df_concat=pd.concat([df_order, df_customers, df_products], axis = 1)
df_concat
```
