```python
#import libraries

import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as npy
df = pd.read_csv('company_sales_data.csv')
print(df)
```

```python
#comparing any two products sales data using a bar chart.

x = npy.arange(len(df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
fig = plt.figure(figsize=(13, 8))
ax = df.plot.bar(x ='month_number', y=['toothpaste', 'facecream'], figsize=(13,7), width=0.6, color=['blue','green'])
plt.legend(loc = 'upper right')
plt.title("kairos aceesories product comparison")
plt.xlabel('Month')
plt.ylabel("sales of facewash and moisturizer")
plt.xticks(x, month)
plt.savefig('Kairos_accessories.png')
plt.show()
```

```python
#Assuming the average monthly sales of bathing soap is 10k, visualize the sales data of bathing soap of all months, showing how they differ from the average sale
fig = plt.figure(figsize=(13,7))
x = npy.arange(len(df['month_number']))
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
yerr = npy.array(list(map(lambda x: 10000-x, df['bathingsoap'])))
yerr = npy.absolute(yerr)

y =df['bathingsoap']
plt.bar(x,y, yerr = yerr, capsize = 11, label = 'bathsoap', color = 'blue')
plt.title('data sales of bathing soap')
plt.xlabel('month')
plt.ylabel('sales of toiletries')
plt.xticks(x, month)
plt.legend()
plt.savefig('data sales of bathing soap.png')
plt.show()
```

```python
#calculating the total sales data for the year of each product and representing it on a pie chart
fig = plt.gcf()
fig.set_size_inches(13,9)
labels = ['Face cream', 'Facewash', 'Toothpaste', 'Bathingsoap', 'Shampoo', 'Moisturizer']
sales_data = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(),
df['shampoo'].sum(), df['moisturizer'].sum()]
plt.pie(sales_data, labels = labels, autopct = '%.d %%', colors = ['black','brown','yellow','red','blue','green'])
plt.title('Total sales data')
plt.axis('equal')
plt.legend(loc = 'upper right')
plt.savefig('Total_sales_data.png')
plt.show()
```
