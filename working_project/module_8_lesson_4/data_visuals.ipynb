```python
import matplotlib
from matplotlib import pyplot as plt 
import pandas as pd
import numpy as npy
import seaborn as sns
```

```python
#loadi the dataset into the dataframe
df = pd.read_csv('WorldCupMatches.csv')
print(df.head())
```

```python
#visualizing the total number of goals scored in each match
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
(df.head())
```

```python
#total number of goals scored in each match
sns.displot(df['Total Goals'], bins = 25, kde = False)
plt.title('Total number of goals in each match')
plt.savefig('Total_number_of_goals_in_each_match.png')
plt.show()
```

```python
#bar chart visualizing how many goals were scored each year the world cup was held betwenn 1930-2014.
sns.set_style('whitegrid')
sns.set_context(rc={'font.size': 13, 'xtick.labelsize': 9})
ax = sns.barplot(data=df, x = 'Year', y = 'Total Goals', errorbar=None, estimator = sum )
plt.xticks(rotation= 'vertical')
ax.set_title('Goals scored each year between 1930-2014')
plt.savefig('Goals_scored_each_year_between_1930-2014.png')
plt.show()
```

```python
#creating a box plot to visualize the distribution of the average goals for eah year with a bar chart.
sns.set_context('notebook', font_scale = 1.27)
f, ax2 = plt.subplots(figsize = (12,7))
palette=sns.color_palette('coolwarm', as_cmap = True)
ax2 = sns.boxplot(data = df, x = 'Year', y = 'Total Goals')
plt.xticks(rotation = 'vertical')
ax2.set_title('Goals per year')
plt.savefig('Goals_per_year.png')
```
